```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset1/merged_file.csv')
df.info()
```

```python
df['DDate'] = pd.to_datetime(df['DDate'])

start_date = '2021-01-01'
end_date = '2024-12-31'

filtered_df = df[
    (df['DDate'] >= start_date) &
    (df['DDate'] <= end_date) &
    (df['CommCode'] == 114) 
    # (df['VarityName'] == 'Common') &
    # (df['YardName'] == 'Bowenpally')
]

filtered_df.head()
```

```python
filtered_df.sort_values(by ='DDate')
filtered_df.head()
```

```python
filtered_df.info()
```

```python
filtered_df_T = filtered_df[ 
     (filtered_df['VarityCode'] == 114) 
]
filtered_df_T.info()
```

```python
filtered_df_T.head(100)
```

```python
# Sort the filtered data by date
filtered_df_T = filtered_df_T.sort_values(by='DDate')

# Display the top rows
filtered_df_T.head()
```

```python
filtered_df_T.tail()
```

```python
filtered_df_T= filtered_df_T.sort_values(by='DDate', ascending=False)
filtered_df_T.head()
```

```python
filtered_df_T = filtered_df_T.sort_values(by='DDate', ascending=True)
filtered_df_T.head()
```

```python
df=filtered_df_T
df.tail()
```

```python
# First ensure DDate is in datetime format and remove time zone if present
df['DDate'] = pd.to_datetime(df['DDate']).dt.date  # Convert to just date part (removes time & timezone)

# Filter as per earlier steps (optional, only if you're working on filtered data)
# df = df[(df['DDate'] >= ...) & (...)]

# Group by DDate, VarityName, and CommName, then compute mean of Model
grouped_df = df.groupby(['DDate', 'VarityName', 'CommName'], as_index=False)['Model'].mean()

# Rename the 'Model' column to indicate it's an average
grouped_df.rename(columns={'Model': 'AvgModel'}, inplace=True)

# Display the result
grouped_df.head()
```
