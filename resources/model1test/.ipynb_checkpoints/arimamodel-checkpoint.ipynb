```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
df = pd.read_excel('2024_2021_tomato_modal_data.xlsx')
```

```python
df
```

```python
# Assuming your DataFrame is df
df['DDate'] = pd.to_datetime(df['DDate'])
df = df.sort_values('DDate')  # Important for time series

# Set date as index
df.set_index('DDate', inplace=True)

# Keep only the target column
ts = df['AvgModel']
```

```python
ts.plot(figsize=(12, 5), title="Average Model Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# Fit ARIMA model
model = ARIMA(ts, order=(5, 1, 2))
model_fit = model.fit()

# Summary of model
print(model_fit.summary())
```

```python
# Split into train/test
train = ts[:-30]
test = ts[-30:]

# Re-fit on train
model = ARIMA(train, order=(5, 1, 2))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=30)

# Evaluation
mse = mean_squared_error(test, forecast)
print(f"Mean Squared Error: {mse:.2f}")

# Plot
plt.figure(figsize=(12, 5))
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, forecast, label="Forecast", linestyle='--')
plt.title("ARIMA Forecast vs Actual")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
